```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

# Step 1: Create sample data (Age, Annual Income)
data = np.array([
    [25, 15000],
    [28, 16000],
    [30, 18000],
    [35, 22000],
    [40, 25000],
    [45, 60000],
    [50, 65000],
    [55, 70000]
])

# Step 2: Standardize the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Step 3: Perform Agglomerative Clustering (Ward linkage, no affinity needed)
model = AgglomerativeClustering(n_clusters=3, linkage='ward')
cluster_labels = model.fit_predict(data_scaled)

# Step 4: Print cluster labels
print("Cluster Labels:", cluster_labels)

# Step 5: Plot the dendrogram (optional)
linked = linkage(data_scaled, method='ward')
plt.figure(figsize=(10, 5))
dendrogram(linked, labels=range(len(data)), color_threshold=0)
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Sample Index")
plt.ylabel("Distance")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 📌 Importing Required Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.preprocessing import StandardScaler

# 📊 Step 1: Create Sample Data (Age, Annual Income)
data = np.array([
    [25, 15000],
    [28, 16000],
    [30, 18000],
    [35, 22000],
    [40, 25000],
    [45, 60000],
    [50, 65000],
    [55, 70000]
])

# 🧼 Step 2: Standardize the Data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# 🌿 Step 3: Apply Hierarchical Clustering (Ward linkage)
model = AgglomerativeClustering(n_clusters=3, linkage='ward')
cluster_labels = model.fit_predict(data_scaled)

# 🖨️ Output cluster labels
print("Cluster Labels:", cluster_labels)

# 🖼️ Step 4: Visualize Clusters
plt.figure(figsize=(8, 5))
plt.scatter(data[:, 0], data[:, 1], c=cluster_labels, cmap='viridis', s=100)
plt.title("Customer Segmentation using Hierarchical Clustering")
plt.xlabel("Age")
plt.ylabel("Annual Income")
plt.grid(True)
plt.show()

# 🌿 Step 5: Plot Dendrogram
linked = linkage(data_scaled, method='ward')

plt.figure(figsize=(10, 6))
dendrogram(linked,
           orientation='top',
           distance_sort='descending',
           show_leaf_counts=True)
plt.title("Dendrogram (Ward Linkage)")
plt.xlabel("Sample Index")
plt.ylabel("Distance")
plt.show()
```
